# DATA SETTLEMENT

```python
sml = pd.concat([pd.read_csv(f,sep='\t').assign(file_name=os.path.basename(f)) for f in sml_files])
sml.columns = sml.columns.str.strip().str.lower()
sml['report_date_local'] =  pd.to_datetime(sml['report_date_local'])
sml['created_date'] =  pd.to_datetime(sml['created_date'])
sml['updated_date'] =  pd.to_datetime(sml['updated_date'])
sml.head()
```

# DATA PROVISION

```python
provision = pd.read_excel(r'C:\Users\Cristian.Aparicio\Documents\PROVISIONES\2023\02. FEBRERO\0. DATA\Detalle_28-02-2023.xlsx',
                          sheet_name='Detalle',
                          converters={'ORDER_ID':str,'LAST4':str,'AFFILIATION':str,'TRX_ID':str,
                                     'DEVICE_ID':str,'ULTIMOS_4':str})
provision.columns = provision.columns.str.strip().str.lower()
provision.head()
```

```python
provision.shape
```

# CRUCE 2: DATA PROVISION VS CRUCE 1

```python
cruce_2 = provision.merge(sml_final[['report_id','applied_fees_benefits','applied_taxes_benefits','fee_direct_deposit',
                                    'taxes_direct_deposit','total_adjustments','total_adjustment_fees_charged',
                                    'total_adjustment_taxes','cobranza','total_released','total_retained','total_due']],
                          on='report_id',how='left')
cruce_2['applied_fees_benefits'].fillna('not_found',inplace=True)
cruce_2['applied_taxes_benefits'].fillna('not_found',inplace=True)
cruce_2['fee_direct_deposit'].fillna('not_found',inplace=True)
cruce_2['taxes_direct_deposit'].fillna('not_found',inplace=True)
cruce_2['total_adjustments'].fillna('not_found',inplace=True)
cruce_2['total_adjustment_fees_charged'].fillna('not_found',inplace=True)
cruce_2['total_adjustment_taxes'].fillna('not_found',inplace=True)
cruce_2['cobranza'].fillna('not_found',inplace=True)
cruce_2['total_released'].fillna('not_found',inplace=True)
cruce_2['total_retained'].fillna('not_found',inplace=True)
cruce_2['total_due'].fillna('not_found',inplace=True)
cruce_2.head()
```

# EXCEL FINAL

```python
writer = pd.ExcelWriter('Data_final_febrero_2023 - ASIGNACION MOVIMIENTOS.xlsx',
                        engine='xlsxwriter',
                        engine_kwargs={'options': {'encoding':'utf-8'}})
cruce_2.to_excel(writer, sheet_name='Detalle', index=False)
writer.close()
```
