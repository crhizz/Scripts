# EDO CUENTA COMPLETO

```python
edo_cta = pd.concat([pd.read_csv(f,sep='|',dtype={'Cuenta': object,'Cod. Transac':object})
                     .assign(file_name=os.path.basename(f)) for f in all_mov])
edo_cta.reset_index(inplace=True)
del edo_cta['index']
edo_cta.sort_values(by=['Movimiento'],inplace=True)
edo_cta['Cuenta'] = "'"+edo_cta['Cuenta']
edo_cta.head()
```

```python
edo_cta.shape
```

```python
split_fecha_operacion = edo_cta["Fecha De Operacin"].str.split("/",n=3,expand=True)
split_fecha_operacion.head()
```

```python
edo_cta["Day_operacion"] = split_fecha_operacion[0] 
edo_cta["Month_operacion"] = split_fecha_operacion[1]
edo_cta['Year_operacion'] = split_fecha_operacion[2]
```

```python
split_fecha_banco = edo_cta["Fecha"].str.split("/",n=3,expand=True)
split_fecha_banco.head()
```

```python
edo_cta["Day_banco"] = split_fecha_banco[0] 
edo_cta["Month_banco"] = split_fecha_banco[1]
edo_cta['Year_banco'] = split_fecha_banco[2]
```

```python
edo_cta['Fecha_operacion'] = edo_cta['Year_operacion']+'-'+edo_cta['Month_operacion']+'-'+edo_cta['Day_operacion']
edo_cta['Fecha_banco'] = edo_cta['Year_banco']+'-'+edo_cta['Month_banco']+'-'+edo_cta['Day_banco']
```

```python
edo_cta['Fecha_operacion'] = pd.to_datetime(edo_cta['Fecha_operacion'])
edo_cta['Fecha_banco'] = pd.to_datetime(edo_cta['Fecha_banco'])
```

```python
edo_cta.loc[edo_cta['Descripcin'].str.contains('08343798'), "Tipo"] = 'XPAY'
edo_cta.loc[edo_cta['Descripcin'].str.contains('08380402'), "Tipo"] = 'MIT-3DS'
edo_cta.loc[(~edo_cta['Descripcin'].str.contains('08343798')) & (~edo_cta['Descripcin'].str.contains('08380402')), "Tipo"] = 'REGULAR TXN'
```

```python
edo_cta.Depsitos = [x.strip('$') for x in edo_cta.Depsitos]
edo_cta.Retiros = [x.strip('$') for x in edo_cta.Retiros]
edo_cta.Saldo = [x.strip('$') for x in edo_cta.Saldo]
```

```python
edo_cta['Depsitos'] = edo_cta['Depsitos'].str.replace(',','').astype(float)
edo_cta['Retiros'] = edo_cta['Retiros'].str.replace(',','').astype(float)
edo_cta['Saldo'] = edo_cta['Saldo'].str.replace(',','').astype(float)
```

```python
edo_cta.loc[edo_cta['Cod. Transac'] == '060', 'Afiliacion_1'] = edo_cta['Descripcin'].str[-9:]
edo_cta.loc[edo_cta['Cod. Transac'] == '061', 'Afiliacion_1'] = edo_cta['Descripcin'].str[-9:]
edo_cta.loc[edo_cta['Cod. Transac'] == '062', 'Afiliacion_1'] = edo_cta['Descripcin'].str[-9:]
edo_cta.loc[edo_cta['Cod. Transac'] == '600', 'Afiliacion_1'] = edo_cta['Descripcin'].str[-9:]
edo_cta.loc[edo_cta['Cod. Transac'] == '601', 'Afiliacion_1'] = edo_cta['Descripcin'].str[-9:]
edo_cta.loc[edo_cta['Cod. Transac'] == '602', 'Afiliacion_1'] = edo_cta['Descripcin'].str[-9:]
edo_cta.loc[(edo_cta['Cod. Transac'] == '602') & (edo_cta['Descripcin Detallada'].str.contains('CONTRACARGO')), 'Afiliacion_1'] = edo_cta['Descripcin'].str[-8:]
edo_cta['len'] = edo_cta['Afiliacion_1'].str.len() 
edo_cta.loc[edo_cta['len'] == 9, 'Afiliacion_final'] = edo_cta['Afiliacion_1'].str[-8:-1]
edo_cta.loc[edo_cta['len'] == 8, 'Afiliacion_final'] = edo_cta['Afiliacion_1'].str[-7:]
edo_cta.loc[edo_cta['len'] == 9, 'Tipo_operacion'] = edo_cta['Afiliacion_1'].str[-1:]
edo_cta.head()
```

```python
edo_cta.shape
```

```python
edo_cta.columns
```

```python
edo_cta_limit = edo_cta[(edo_cta['Fecha_operacion'] >= start_date) & (edo_cta['Fecha_operacion'] <= final_date)]
edo_cta_limit['Fecha_operacion'].unique()
```

```python
edo_cta_final = edo_cta_limit[['Cuenta','Fecha_operacion','Fecha_banco','Referencia','Descripcin','Cod. Transac','Sucursal',
                               'Depsitos','Retiros','Saldo','Movimiento','Descripcin Detallada','Cheque','Afiliacion_final',
                              'Tipo_operacion']]
edo_cta_final['Afiliacion_final'].fillna('not_found',inplace=True)
edo_cta_final['Tipo_operacion'].fillna('not_found',inplace=True)
```

```python
edo_cta_final.head()
```

```python
edo_cta_final.shape
```

```python
edo_cta_final_deb_cred = edo_cta_final[(edo_cta_final['Tipo_operacion'] != 'not_found')]
edo_cta_final_deb_cred.head()
```

```python
edo_cta_final_deb_cred.shape
```

```python
edo_cta_final_deb_cred['Tipo_operacion'].unique()
```

```python
pt = pd.pivot_table(edo_cta_final_deb_cred,index=['Fecha_operacion'],
                    columns=['Tipo_operacion'],
                    values=['Depsitos','Retiros']
                    ,aggfunc=np.sum,margins=True)
pt = pt.reset_index(drop=False)
# pt.to_csv('PIVOT BY EMAIL ABRIL.csv')
pt.head()
```
