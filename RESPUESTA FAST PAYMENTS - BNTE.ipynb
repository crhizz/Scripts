# ARCHIVO DISPERSION

```python
disp1.head()
```

```python
disp1['report_date_local'] = pd.to_datetime(disp1['report_date_local'])
```

```python
split = disp1["swap"].str.split(",",n=7,expand=True)
split.head()
```

```python
disp1["swap"] = split[0] 
disp1["CLABE"] = split[3]
```

```python
disp1.head()
```

```python
disp1.loc[disp1['report_date_local'] == date_respuesta, 'tipo'] = 'Pago del da'
disp1.loc[disp1['report_date_local'] != date_respuesta, 'tipo'] = 'Reprogramacin'
disp1.drop(['bank_name','CLABE','status'],axis=1,inplace=True)
disp1.head()
```

```python
disp_2 = disp1.merge(rs_concat_2[['INDEX','status','Nombre de la institucin','CLABE','LINEA CEP','Time_rs','Banco_emisor']]
                     ,on='INDEX', how='left')
disp_2.sort_values(by=['status'])
disp_2["Time"].fillna('NO TIME',inplace=True)
disp_2.loc[disp_2['Time'] == 'NO TIME', 'status'] = 'ERROR'
disp_2.loc[(disp_2['LINEA CEP'].str.contains('nan')) & (['Nombre de la institucin'] == 'BANORTE/IXE'), 'status'] = 'ERROR'
# disp_2.loc[pd.notnull(disp_2['error_description']), 'status'] = 'ERROR'
disp_2.head()
```

```python
disp_2.loc[disp_2['Nombre de la institucin'] == 'BANORTE/IXE', 'LINEA CEP'] = 'COMPROBANTE MISMO BANCO BANORTE'
disp_2.loc[disp_2['status'] == 'ERROR', 'LINEA CEP'] = ''
```

```python
for col in disp_2.columns: 
    print(col) 
```

```python
disp_final = disp_2[['merchant_id','merchant_name','user_email','report_id','report_date_local','amount','tipo'
                    ,'Nombre de la institucin','swap','CLABE','status','LINEA CEP','Time','Banco_emisor']]
disp_final.rename(columns ={'Nombre de la institucin':'bank','amount':'total_due','LINEA CEP':'CEP'}, inplace=True)
disp_final.sort_values(by=['status','Time'], ascending=True, inplace=True)
disp_final.drop_duplicates(subset =['report_id','Time'], keep ='first', inplace=True)
disp_final.head()
```

```python
disp_final.shape
```
