# DATA EDO CTA

### CSV

```python
edo_cta_1 = pd.concat([pd.read_csv(f).assign(file_name=os.path.basename(f)) for f in all_mov_csv])
edo_cta_1.reset_index(inplace=True)
edo_cta_1.head()
```

```python
edo_cta_1.shape
```

### XLSX

```python
edo_cta_2 = pd.concat([pd.read_excel(f).assign(file_name=os.path.basename(f)) for f in all_mov_xlsx])
edo_cta_2.reset_index(inplace=True)
edo_cta_2.head()
```

### CONSOLIDADO 

```python
edo_cta = pd.concat([edo_cta_1,edo_cta_2])
edo_cta.head()
```

```python
edo_cta.shape
```

```python
edo_cta_all = edo_cta [['Bank Account Number','Operation Date','Value Date','Bank Reference','Description','Debit/Credit Mark',
                       'Credit','Debit','Amount','CCY','Budget Code','Customer Reference','file_name']]
edo_cta_all['Bank Reference'].fillna('blank_spaces',inplace=True)
edo_cta_all.head()
```
