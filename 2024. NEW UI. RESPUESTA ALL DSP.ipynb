```python
rs1['report_id'] = rs1['report_id'].str.strip()
```

```python
rs2 = rs1.merge(bin1[['BIN','Clave de la instituci√≥n','Nombre de la instituci√≥n']] ,on='BIN', how='left')
rs2['Comision'] = ''
rs2['LINEA CEP'] = (rs2['Date_respuesta'].astype(str).replace(' ', '',regex=True)+','
                    +rs2['disbursement_tracking_key'].astype(str)+','
                    +rs2['Banco emisor'].astype(str).replace(' ', '',regex=True)+','
                    +rs2['Clave de la instituci√≥n'].astype(str).replace(' ', '',regex=True)+','
                    +rs2['CLABE'].astype(str).replace(' ', '',regex=True)+','
                    +rs2['amount'].astype(str).replace(' ', '',regex=True))
rs2['Metodo de dispersion'] = rs2['Banco_emisor']
rs2.head()
```

```python
rs2.shape
```

# ARCHIVO DISPERSION

```python
disp['date_disbursement'] = date_respuesta
disp['report_id'] = disp['report_id'].str.strip()
disp['report_date_local'] = pd.to_datetime(disp['report_date_local'])
```

```python
disp.loc[disp['report_date_local'] == date_respuesta, 'tipo'] = 'Pago del d√≠a'
disp.loc[disp['report_date_local'] != date_respuesta, 'tipo'] = 'Reprogramaci√≥n'
disp.drop(['bank_name','bank_account_number','status','swap','payment_date_local'],axis=1,inplace=True)
disp.head()
```

```python
disp.shape
```

```python
disp_2 = disp.merge(rs2[['INDEX','status','Nombre de la instituci√≥n','CLABE','LINEA CEP','Banco_emisor']]
                    ,on='INDEX', how='left')
disp_2.loc[disp_2['status'] != 'PAID', 'status'] = 'ERROR'
disp_2.sort_values(by=['status'])
disp_2.head()
```

```python
disp_2.loc[((disp_2['Banco_emisor'] == 'BANAMEX') &
            (disp_2['Nombre de la instituci√≥n'] == 'BANAMEX')), 'LINEA CEP'] = 'COMPROBANTE MISMO BANCO BANAMEX'
disp_2.loc[((disp_2['Banco_emisor'] == 'SANTANDER') &
            (disp_2['Nombre de la instituci√≥n'] == 'SANTANDER')), 'LINEA CEP'] = 'COMPROBANTE MISMO BANCO SANTANDER'
disp_2.loc[((disp_2['Banco_emisor'] == 'BANORTE') &
            (disp_2['Nombre de la instituci√≥n'] == 'BANORTE/IXE')), 'LINEA CEP'] = 'COMPROBANTE MISMO BANCO BANORTE'
disp_2.loc[disp_2['status'] == 'ERROR', 'LINEA CEP'] = ''
```

```python
for col in disp_2.columns: 
    print(col) 
```

```python
disp_final = disp_2[['type','merchant_id','merchant_name','user_email','report_id','report_date_local','amount','tipo'
                    ,'Nombre de la instituci√≥n','bank_account_owner_name','CLABE','status','LINEA CEP','Batch_time',
                     'Banco_emisor','INDEX']]
disp_final.rename(columns ={'Nombre de la instituci√≥n':'bank','amount':'total_due'}, inplace=True)
disp_final.sort_values(by=['status','Batch_time'], ascending=True, inplace=True)
disp_final.head()
```
