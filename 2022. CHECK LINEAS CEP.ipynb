```python
import pandas as pd
import numpy as np
from openpyxl import load_workbook
import glob
import os
pd.options.display.float_format = '{:,.2f}'.format
```

### 2022

```python
allfiles_disp_trad_may_2022 = sorted(glob.glob(r'C:\Users\Cristian.Aparicio\Documents\DISPERSIONES\2022\**\ARCHIVO DISPERSIONES\Dispersión*.xlsx', recursive=False))
allfiles_disp_trad_may_2022
```

```python
allfiles_fp_may_2022 = sorted(glob.glob(r'C:\Users\Cristian.Aparicio\Documents\FAST PAYMENTS\2022\**\ARCHIVO DISPERSIONES FP\Dispersión-FP*.xlsx',recursive=False))
allfiles_fp_may_2022
```

```python
swap_dia_total_mayo_2022 = pd.concat([pd.read_excel(f,sheet_name = 'Swap del día')
                            .assign(file_name=os.path.basename(f)) for f in allfiles_disp_trad_may_2022])
swap_dia_total_mayo_2022['Date'] = swap_dia_total_mayo_2022["file_name"].str[-13:-5]
swap_dia_total_mayo_2022['Date']= pd.to_datetime(swap_dia_total_mayo_2022['Date'], format='%Y%m%d').dt.date
swap_dia_total_mayo_2022['Date']= pd.to_datetime(swap_dia_total_mayo_2022['Date'])
swap_dia_total_mayo_2022.sort_values(by=['Date'],ascending=True,inplace=True)
swap_dia_total_mayo_2022['report_date_local']= pd.to_datetime(swap_dia_total_mayo_2022['report_date_local'])
swap_dia_total_mayo_2022.head()
```

```python
swap_dia_total_mayo_2022.shape
```

```python
fp_dia_total_mayo_2022 = pd.concat([pd.read_excel(f,sheet_name='FAST PAYMENTS')
                          .assign(file_name=os.path.basename(f)) for f in allfiles_fp_may_2022])
fp_dia_total_mayo_2022['Date'] = fp_dia_total_mayo_2022["file_name"].str[-15:-5]
fp_dia_total_mayo_2022['Date']= pd.to_datetime(fp_dia_total_mayo_2022['Date'])
fp_dia_total_mayo_2022['report_date_local']= pd.to_datetime(fp_dia_total_mayo_2022['report_date_local'])
fp_dia_total_mayo_2022.head()
```

### ALL FILES

```python
allfiles = pd.concat([swap_dia_total_mayo_2022,fp_dia_total_mayo_2022])
allfiles.fillna('BLANK',inplace=True)
allfiles['Date']= pd.to_datetime(allfiles['Date'])
allfiles.sort_values(by=['Date','report_id'],inplace=True)
allfiles.drop(['status'],axis=1,inplace=True)
allfiles.head()
```
