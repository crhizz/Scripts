### LECTURA DE ARCHIVOS

```python
swap_dia_total_jan_2023 = pd.concat([pd.read_excel(f,sheet_name = 'Swap del día')
                            .assign(file_name=os.path.basename(f)) for f in allfiles_disp_trad_jan_2023])
swap_dia_total_jan_2023['Date'] = swap_dia_total_jan_2023["file_name"].str[-13:-5]
swap_dia_total_jan_2023['Date']= pd.to_datetime(swap_dia_total_jan_2023['Date'], format='%Y%m%d').dt.date
swap_dia_total_jan_2023['Date']= pd.to_datetime(swap_dia_total_jan_2023['Date'])
swap_dia_total_jan_2023.sort_values(by=['Date'],ascending=True,inplace=True)
swap_dia_total_jan_2023['report_date_local']= pd.to_datetime(swap_dia_total_jan_2023['report_date_local'])
swap_dia_total_jan_2023.head()
```
