# DATA AP JOURNAL

```python
ap = pd.concat([pd.read_csv(f, sep='\t').assign(file_name=os.path.basename(f)) for f in ap_files])
ap.columns = ap.columns.str.strip().str.lower()
ap.replace('""','', regex=True, inplace=True)
ap['eff_dt'] =  pd.to_datetime(ap['eff_dt']).dt.tz_localize(None)
ap['eff_dt_date'] =  pd.to_datetime(ap['eff_dt']).dt.date
ap['eff_dt_date'] =  pd.to_datetime(ap['eff_dt_date'])
ap['eff_dt_time'] =  pd.to_datetime(ap['eff_dt']).dt.time
ap['movement_day'] = pd.to_datetime(ap['movement_day']).dt.tz_localize(None)
ap['created_at'] = pd.to_datetime(ap['created_at']).dt.tz_localize(None)
ap['created_at_date'] =  pd.to_datetime(ap['created_at']).dt.date
ap['created_at_date'] =  pd.to_datetime(ap['created_at_date'])
ap['request_effective_date'] = pd.to_datetime(ap['request_effective_date']).dt.tz_localize(None)
ap['slot_date'] = pd.to_datetime(ap['slot_date']).dt.tz_localize(None)
ap['identifier'] = ap['source_type'].str[:1].astype(str)
# ap['INDEX'] = (ap['eff_dt_date'].astype(str).replace(' ', '',regex=True)+'-'+
#                ap['identifier'].astype(str).replace(' ', '',regex=True)+'-'+
#                ap['eff_dt_time'].astype(str).replace(' ', '',regex=True))
ap.sort_values(by=(['merchant_id','eff_dt']), ascending=[True,False], inplace=True)
ap.head()
```

## CRUCE 1: SML VS AP JOURNAL - TRANSACCIONES

```python
cruce_1 = sml_mnf.merge(ap_mnf_sett[['mvnt_id','report_id','eff_dt_date','eff_dt_time','movement_type']],on='mvnt_id',how='left')
cruce_1['report_id'].fillna('not_found',inplace=True)
cruce_1.rename(columns={'psp_amount':'psp_amount_sett','gross_amount':'gross_amount_sett','fees':'fees_sett',
                                'taxes':'taxes_sett','net_amount':'net_amount_sett'},inplace=True)
cruce_1.head()
```

```python
cruce_1['movement_type'].unique()
```

## CRUCE 2: SML VS AP JOURNAL - CANCELACIONES

```python
cruce_2 = sml_mnf.merge(ap_mnf_canc[['mvnt_id','report_id','eff_dt_date','eff_dt_time','movement_type']],on='mvnt_id',how='left')
cruce_2['report_id'].fillna('not_found',inplace=True)
cruce_2['movement_type'].fillna('not_found',inplace=True)
cruce_2.head()
```

```python
cruce_2.shape
```

```python
cruce_2['movement_type'].unique()
```

```python
cruce_2_canc = cruce_2[(cruce_2['movement_type'] == 'CANCEL_RQ')]
cruce_2_canc.rename(columns={'psp_amount':'psp_amount_pymt','gross_amount':'gross_amount_pymt','fees':'fees_pymt',
                             'taxes':'taxes_pymt','net_amount':'net_amount_pymt'},inplace=True)
cruce_2_canc.head()
```

```python
cruce_2_canc.shape
```

## CRUCE FINAL: RESTAR CANCELACIONES DE TRANSACCIONES

```python
cruce_3 = cruce_1.merge(cruce_2_canc[['mvnt_id','psp_amount_pymt','gross_amount_pymt','fees_pymt','taxes_pymt',
                                                             'net_amount_pymt']],on='mvnt_id',how='left')
cruce_3.fillna(0,inplace=True)
cruce_3['psp_amount'] = cruce_3['psp_amount_sett'] - cruce_3['psp_amount_pymt'] 
cruce_3['gross_amount'] = cruce_3['gross_amount_sett'] - cruce_3['gross_amount_pymt'] 
cruce_3['fees'] = cruce_3['fees_sett'] - cruce_3['fees_pymt'] 
cruce_3['taxes'] = cruce_3['taxes_sett'] - cruce_3['taxes_pymt'] 
cruce_3['net_amount'] = cruce_3['net_amount_sett'] - cruce_3['net_amount_pymt']
cruce_3.head()
```

```python
cruce_3.shape
```

```python
cruce_3_report_id_found = cruce_3[(cruce_3['report_id'] != 'not_found')]
cruce_3_report_id_found.shape
```

```python
cruce_3_report_id_not_found = cruce_3[(cruce_3['report_id'] == 'not_found')]
cruce_3_report_id_not_found.shape
```
