# DATA SETTLEMENT

```python
sml = pd.concat([pd.read_csv(f,sep='\t').assign(file_name=os.path.basename(f)) for f in sml_files])
sml.columns = sml.columns.str.strip().str.lower()
sml['report_date_local'] =  pd.to_datetime(sml['report_date_local'])
sml['payment_date_local'] =  pd.to_datetime(sml['payment_date_local']).dt.tz_localize(None)
sml['payment_date_local_final'] =  pd.to_datetime(sml['payment_date_local']).dt.date
sml['payment_date_local_final'] =  pd.to_datetime(sml['payment_date_local_final'])
sml.head()
```

# DATA PROVISION

```python
provision = pd.read_excel(r'C:\Users\Cristian.Aparicio\Documents\PROVISIONES\2023\02. FEBRERO\0. DATA\Detalle_28-02-2023.xlsx',
                          sheet_name='Detalle',
                          converters={'ORDER_ID':str,'LAST4':str,'AFFILIATION':str,'TRX_ID':str,
                                     'DEVICE_ID':str,'ULTIMOS_4':str})
provision.columns = provision.columns.str.strip().str.lower()
provision.head()
```

# CRUCE 2: DATA PROVISION VS CRUCE 1

```python
cruce_2 = provision.merge(sml_final[['transaction_id','report_id','report_date_local','payment_date_local',
                                    'payment_date_local_final','status_disbursement','fees_charged','taxes']],
                          on='transaction_id',how='left')
cruce_2['report_id'].fillna('not_found',inplace=True)
cruce_2['report_date_local'].fillna('1990-01-01',inplace=True)
cruce_2['payment_date_local'].fillna('1990-01-01',inplace=True)
cruce_2['report_date_local'] =  pd.to_datetime(cruce_2['report_date_local'])
cruce_2['payment_date_local_final'] =  pd.to_datetime(cruce_2['payment_date_local_final'])
cruce_2['fees_charged'].fillna('not_found',inplace=True)
cruce_2['taxes'].fillna('not_found',inplace=True)
cruce_2.head()
```

# EXCEL FINAL

```python
writer = pd.ExcelWriter('Data_final_febrero_2023 - FEES TAXES.xlsx',
                        engine='xlsxwriter',
                        engine_kwargs={'options': {'encoding':'utf-8'}})
cruce_2.to_excel(writer, sheet_name='Detalle', index=False)
writer.close()
```
