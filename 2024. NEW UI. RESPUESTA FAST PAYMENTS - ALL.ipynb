## QUERY PAYMENTS

```python
query = pd.read_csv(r'G:\.shortcut-targets-by-id\1Oe28Mf9uSidMamjllDen2gkNnC59eSUr\3.-Settlement\5.-HISTORICO FILES PAGOS\02. FAST PAYMENTS\2024\01. ENERO\FP-ALL 2024-01-31.csv')
query.columns = query.columns.str.strip().str.lower() 
query.loc[query['type'].isnull(), 'type'] = 'LEDGER_FP'
query['batch_created_date'] = pd.to_datetime(query['batch_created_date']).dt.tz_localize(None)
query['payment_date_local'] = pd.to_datetime(query['payment_date_local']).dt.tz_localize(None)
query.set_index('batch_created_date',inplace=True)
query.head()
```

# ARCHIVO DISPERSION

```python
disp1.head()
```

```python
disp1['report_date_local'] = pd.to_datetime(disp1['report_date_local'])
```

```python
split = disp1["swap"].str.split(",",n=7,expand=True)
split.head()
```

```python
disp1["swap"] = split[0] 
disp1["CLABE"] = split[3]
```

```python
disp1.head()
```

```python
disp1.loc[disp1['report_date_local'] == date_respuesta, 'tipo'] = 'Pago del día'
disp1.loc[disp1['report_date_local'] != date_respuesta, 'tipo'] = 'Reprogramación'
disp1.drop(['bank_name','CLABE','status'],axis=1,inplace=True)
disp1.head()
```

```python
disp_2 = disp1.merge(rs_concat_2[['INDEX','status','Nombre de la institución','CLABE','LINEA CEP','Batch_time','Banco_emisor']]
                     ,on='INDEX', how='left')
disp_2.sort_values(by=['status'])
disp_2["Batch_time"].fillna('NO TIME',inplace=True)
disp_2.loc[disp_2['Batch_time'] == 'NO TIME', 'status'] = 'ERROR'
disp_2.loc[(disp_2['LINEA CEP'].str.contains('nan')) & (['Nombre de la institución'] == 'BANAMEX'), 'status'] = 'ERROR'
disp_2.loc[disp_2['status'] != 'PAID', 'status'] = 'ERROR'
disp_2.head()
```

```python
disp_2.loc[(disp_2['Nombre de la institución'] == 'BANAMEX') & (disp_2['Banco_emisor'] == 'BANAMEX'),
           'LINEA CEP'] = 'COMPROBANTE MISMO BANCO BANAMEX'
disp_2.loc[(disp_2['Nombre de la institución'] == 'SANTANDER') & (disp_2['Banco_emisor'] == 'SANTANDER'),
           'LINEA CEP'] = 'COMPROBANTE MISMO BANCO SANTANDER'
disp_2.loc[(disp_2['Nombre de la institución'] == 'BANORTE/IXE') & (disp_2['Banco_emisor'] == 'BANORTE'),
           'LINEA CEP'] = 'COMPROBANTE MISMO BANCO BANORTE'
disp_2.loc[disp_2['status'] == 'ERROR', 'LINEA CEP'] = ''
```

```python
for col in disp_2.columns: 
    print(col) 
```

```python
disp_final = disp_2[['type','merchant_id','merchant_name','user_email','report_id','report_date_local','amount','tipo'
                    ,'Nombre de la institución','swap','CLABE','status','LINEA CEP','Batch_time','Banco_emisor','INDEX']]
disp_final.rename(columns ={'Nombre de la institución':'bank','amount':'total_due'}, inplace=True)
disp_final.sort_values(by=['status','Batch_time'], ascending=True, inplace=True)
# disp_final.drop_duplicates(subset =['report_id','Time'], keep ='first', inplace=True)
disp_final.head()
```
