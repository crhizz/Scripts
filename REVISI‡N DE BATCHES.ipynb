```python
import pandas as pd
import numpy as np
import glob
import os
from openpyxl import load_workbook
pd.options.display.float_format = '{:,.2f}'.format
```

### ARCHIVOS DE DISPERSIÓN TRADICIONAL

```python
allfiles_disp_trad = glob.glob(r'C:\Users\Cristian.Aparicio\Documents\DISPERSIONES\2022\09. SEPTIEMBRE\ARCHIVO DISPERSIONES\Dispersión*.xlsx', recursive=False)
allfiles_disp_trad
```

```python
swap_dia_total = pd.concat([pd.read_excel(f,sheet_name = 'Swap del día')
                            .assign(file_name=os.path.basename(f)) for f in allfiles_disp_trad])
swap_dia_total['Date'] = swap_dia_total["file_name"].str[-13:-5]
swap_dia_total['Date']= pd.to_datetime(swap_dia_total['Date'], format='%Y%m%d').dt.date
swap_dia_total['Date']= pd.to_datetime(swap_dia_total['Date'])
swap_dia_total.sort_values(by=['Date'],ascending=True,inplace=True)
swap_dia_total['report_date_local']= pd.to_datetime(swap_dia_total['report_date_local'])
swap_dia_total['INDEX'] = (swap_dia_total['report_id'].astype(str).replace(' ', '',regex=True) + '_' +
                           swap_dia_total['Date'].astype(str).replace(' ', '',regex=True))
                           
swap_dia_total.head()
```

### SEPARACIÓN DE BATCHES

```python
swap_dia_first_batch = swap_dia_total[(swap_dia_total['Batch_time'] == 'First_batch')]
swap_dia_first_batch.head()
```

```python
swap_dia_first_batch.shape
```

```python
duplicates_first_batch = swap_dia_first_batch[swap_dia_first_batch.duplicated(['INDEX'] , keep=False)]
duplicates_first_batch.head()
```

```python
duplicates_first_batch.shape
```

```python
swap_dia_second_batch = swap_dia_total[(swap_dia_total['Batch_time'] == 'Second_batch')]
swap_dia_second_batch.head()
```

```python
swap_dia_second_batch.shape
```

```python
duplicates_second_batch = swap_dia_second_batch[swap_dia_second_batch.duplicated(['INDEX'] , keep=False)]
duplicates_second_batch.head()
```

```python
duplicates_second_batch.shape
```

### CRUCE SEGUNDO BATCH VS. PRIMER BATCH 

```python
revision_1 = swap_dia_second_batch.merge(swap_dia_first_batch[['INDEX','Batch_time']] ,on='INDEX', how='left')
revision_1['Batch_time_y'].fillna('not_found',inplace=True)
revision_1.head()
```

```python
revision_1.shape
```

```python
writer = pd.ExcelWriter('Revision_batches_septiembre_2022.xlsx',
                        engine='xlsxwriter',
                        engine_kwargs={'options': {'encoding':'utf-8'}})
swap_dia_second_batch.to_excel(writer, sheet_name='Segundo batch',index=False)
revision_1.to_excel(writer, sheet_name='Validacion con primer batch',index=False)
writer.close()
```
