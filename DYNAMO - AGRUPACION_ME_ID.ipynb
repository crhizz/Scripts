# DATA SETTLEMENT MASTER LOG

```python
sml = pd.concat([pd.read_csv(f).assign(file_name=os.path.basename(f)) for f in sml_files])
sml.columns = sml.columns.str.strip().str.lower()
sml.replace('""','', regex=True, inplace=True)
sml['created_at'] = pd.to_datetime(sml['created_at']).dt.tz_localize(None)
sml['created_at_date'] =  pd.to_datetime(sml['created_at']).dt.date
sml['created_at_date'] =  pd.to_datetime(sml['created_at_date'])
sml['eff_dt'] = pd.to_datetime(sml['eff_dt']).dt.tz_localize(None)
sml['eff_dt_date'] =  pd.to_datetime(sml['eff_dt']).dt.date
sml['eff_dt_date'] =  pd.to_datetime(sml['eff_dt_date'])
sml.rename(columns={'ref_id':'transaction_id'},inplace=True)
#sml['PK'] = sml['PK'].replace('M#','',inplace=True)
sml.head()
```

```python
sml.shape
```

```python
sml_final = sml[(sml['mvnt_type'] == 'STL_PCS_RQ')]
sml_final.head()
```

# DATA AP JOURNAL

```python
ap = pd.concat([pd.read_csv(f).assign(file_name=os.path.basename(f)) for f in ap_files])
ap.columns = ap.columns.str.strip().str.lower()
ap.replace('""','', regex=True, inplace=True)
ap['eff_dt'] =  pd.to_datetime(ap['eff_dt']).dt.tz_localize(None)
ap['eff_dt_date'] =  pd.to_datetime(ap['eff_dt']).dt.date
ap['eff_dt_date'] =  pd.to_datetime(ap['eff_dt_date'])
ap['eff_dt_time'] =  pd.to_datetime(ap['eff_dt']).dt.time
ap['movement_day'] = pd.to_datetime(ap['movement_day']).dt.tz_localize(None)
ap['created_at'] = pd.to_datetime(ap['created_at']).dt.tz_localize(None)
ap['created_at_date'] =  pd.to_datetime(ap['created_at']).dt.date
ap['created_at_date'] =  pd.to_datetime(ap['created_at_date'])
ap['request_effective_date'] = pd.to_datetime(ap['request_effective_date']).dt.tz_localize(None)
ap['slot_date'] = pd.to_datetime(ap['slot_date']).dt.tz_localize(None)
ap['identifier'] = ap['source_type'].str[:1].astype(str)
ap.sort_values(by=(['merchant_id','eff_dt']), ascending=[True,False], inplace=True)
```

```python
ap['INDEX'] = (ap['eff_dt_date'].astype(str).replace(' ', '',regex=True)+'-'+
               ap['identifier'].astype(str).replace(' ', '',regex=True)+'-'+
               ap['eff_dt_time'].astype(str).replace(' ', '',regex=True))
```

```python
ap_final = ap.reset_index(drop=True)
ap_final.head()
```

```python
ap_final.loc[ap_final['source_type'] == 'DISB', 'report_id'] = ap_final['type_request']
ap_final['report_id'] = ap_final['report_id'].groupby(ap_final['merchant_id']).fillna(method='ffill')
ap_final.head()
```

```python
ap_final.shape
```

```python
ap_final.dtypes
```

```python
ap_final['source_type'].unique()
```

# CRUCE_1: AP JOURNAL VS SETTLEMENT MASTER LOG

```python
cruce_1 = ap_final.merge(sml_final[['mvnt_id','gross_amount','fees','taxes','net_amount','mdr','msi']],
                            on='mvnt_id',how='left')
cruce_1.head()
```

```python
cruce_1.shape
```

# PERIODO FACTURACION

```python
cruce_1_period = cruce_1[(cruce_1['eff_dt_date'] >= start_date_fact) & (cruce_1['eff_dt_date'] <= final_date_fact)]
cruce_1_period.head()
```

```python
cruce_1_period.shape
```

```python
sorted(cruce_1_period['eff_dt_date'].unique())
```

```python
ap_sett = cruce_1_period[(cruce_1_period['source_type'] == 'SETT')]
ap_topups = cruce_1_period[(cruce_1_period['source_type'] == 'TOPUPS')]
ap_disb = cruce_1_period[(cruce_1_period['source_type'] == 'DISB')]
ap_loan = cruce_1_period[(cruce_1_period['source_type'] == 'LOAN')]
ap_pymt = cruce_1_period[(cruce_1_period['source_type'] == 'PYMT')]
ap_risk = cruce_1_period[(cruce_1_period['source_type'] == 'RISK')]
```

### TPV, FEES, TAXES, NET AMOUNT

```python
cruce_1_group = ap_sett.groupby(['eff_dt_date','merchant_id','report_id']).agg({'mvnt_id':'count',
                                                                      'gross_amount':sum,
                                                                      'fees':sum,
                                                                      'taxes':sum,
                                                                      'net_amount':sum,
                                                                      'amount':sum}).reset_index()
cruce_1_group.head()
```

```python
cruce_1_group.shape
```

```python
sorted(cruce_1_group['eff_dt_date'].unique())
```

```python
cruce_1_group_me_id = ap_sett.groupby(['merchant_id','eff_dt_date','report_id']).agg({'mvnt_id':'count',
                                                                      'gross_amount':sum,
                                                                      'fees':sum,
                                                                      'taxes':sum,
                                                                      'net_amount':sum,
                                                                      'amount':sum}).reset_index()
cruce_1_group_me_id.head()
```

```python
cruce_1_group_me_id.shape
```

### TOPUPS

```python
ap_topups['fee_direct_deposit'] = ap_topups['amount'] / 1.16
ap_topups['taxes_direct_deposit'] = ap_topups['fee_direct_deposit'] * 0.16
ap_topups.head()
```

```python
ap_topups.shape
```

```python
ap_topups_group_me_id = ap_topups.groupby(['report_id']).agg({'mvnt_id':'count',
                                                     'fee_direct_deposit':sum,
                                                     'taxes_direct_deposit':sum,
                                                     'amount':sum}).reset_index()
ap_topups_group_me_id.head()
```

```python
ap_topups_group_me_id.shape
```

### LOANS

```python
ap_loan.head()
```

```python
ap_loan.shape
```

```python
ap_loan_group_me_id = ap_loan.groupby(['report_id']).agg({'mvnt_id':'count',
                                                     'amount':sum}).reset_index()
ap_loan_group_me_id.rename(columns={'amount':'amount_loans'},inplace=True)
ap_loan_group_me_id.head()
```

### RISK 

```python
ap_risk.head()
ap_risk.rename(columns={'amount':'amount_risk'},inplace=True)
```

```python
ap_risk.shape
```

```python
ap_risk_group_me_id = pd.pivot_table(ap_risk,index=['report_id'],
                                  columns=['movement_type'],
                                  values=['amount_risk']
                                  ,aggfunc = ['sum'],margins=True).reset_index()
ap_risk_group_me_id.fillna(0,inplace=True)
ap_risk_group_me_id.head()
```

```python
ap_risk_group_me_id.shape
```

```python
ap_risk_group_me_id.columns = ['_'.join(col) for col in ap_risk_group_me_id.columns.values]
ap_risk_group_me_id.rename(columns={'report_id__':'report_id',
                                   'sum_amount_risk_CHRBCK_RQ':'chargeback_rq',
                                   'sum_amount_risk_DISPUTE_ADJUSTMENT_RQ':'dispute_adjustment_rq',
                                   'sum_amount_risk_DISPUTE_RQ':'risk_dispute_rq',
                                   'sum_amount_risk_RELEASE_RQ':'risk_release_rq',
                                   'sum_amount_risk_REVIEW_RQ':'risk_review_rq',
                                   'sum_amount_risk_All':'risk_all'},inplace=True)
ap_risk_group_me_id.head()
```

```python
ap_risk_group_me_id.shape
```

```python
ap_risk_group_me_id.columns
```

### DISBURSEMENT

```python
ap_disb.head()
ap_disb.rename(columns={'amount':'amount_disbursement'},inplace=True)
```

```python
ap_disb.shape
```

```python
ap_disb_group_me_id = pd.pivot_table(ap_disb,index=['report_id'],
                                  columns=['movement_type'],
                                  values=['amount_disbursement']
                                  ,aggfunc = ['sum'],margins=True).reset_index()
ap_disb_group_me_id.fillna(0,inplace=True)
ap_disb_group_me_id.head()
```

# UNION_1: SETT + TOPUPS

```python
union_1 = cruce_1_group_me_id.merge(ap_topups_group_me_id[['report_id','fee_direct_deposit','taxes_direct_deposit']],
                            on='report_id',how='left')
union_1.fillna(0,inplace=True)
union_1.head()
```

```python
union_1.shape
```

# UNION_2: UNION_1 + LOANS

```python
union_2 = union_1.merge(ap_loan_group_me_id[['report_id','amount_loans']],
                            on='report_id',how='left')
union_2.fillna(0,inplace=True)
union_2.head()
```

```python
union_2.shape
```

# UNION_3: UNION_2 + RISK

```python
union_3 = union_2.merge(ap_risk_group_me_id[['report_id','chargeback_rq','dispute_adjustment_rq','risk_dispute_rq',
                                             'risk_release_rq','risk_review_rq']],on='report_id',how='left')
union_3.fillna(0,inplace=True)
union_3.head()
```

```python
union_3.shape
```

# UNION_4: UNION_3 + CANCELACIONES

```python
union_4 = union_3.merge(ap_pymt_me_id[['report_id','gross_amount_canc','fees_canc','taxes_canc','net_amount_canc',
                                       'amount_canc']],on='report_id',how='left')
union_4.fillna(0,inplace=True)
union_4.head()
```

```python
union_4.shape
```

# UNION_5: UNION_4 + DISBURSEMENT

```python
union_5 = union_4.merge(ap_disb_group_me_id[['report_id','disbursement_rq','disbursement_return_rq']]
                        ,on='report_id',how='left')
union_5.fillna(0,inplace=True)
union_5.head()
```
