```python
all_respuestas_notskiprows = pd.concat([pd.read_csv(f).assign(file_name=os.path.basename(f)) for f in respuestas_swap_notskiprows_files])
all_respuestas_notskiprows.head()
```

### CONSOLIDADO DE RESPUESTAS

```python
all_respuestas = pd.concat([all_respuestas_notskiprows])
all_respuestas['Metodo de dispersion'].fillna('SWAP_TENT',inplace=True)
all_respuestas.head()
```

# SWAP  

```python
swap = all_respuestas[(all_respuestas['Metodo de dispersion'] == 'SWAP')]
```

### SWAP ALL

```python
swap_all = all_respuestas.groupby(['Metodo de dispersion','file_name']).agg({'Importe':sum,'report_id':'count'}).reset_index()
swap_all['COMISION'] = 1.85
swap_all['IVA'] = swap_all['COMISION'] * 0.16
swap_all['COMISION + IVA'] = swap_all['COMISION'] + swap_all['IVA']
swap_all['COMISION_FINAL'] = swap_all['COMISION'] * swap_all['report_id']
swap_all['IVA_FINAL'] = swap_all['COMISION_FINAL'] * 0.16
swap_all['COMISION_CALCULADA'] = swap_all['COMISION_FINAL'] + swap_all['IVA_FINAL']
swap_all['COMISION_TOTAL'] = swap_all['report_id'] * swap_all['COMISION + IVA']
swap_all['TOTAL_DEPOSITADO'] = swap_all['Importe'] + swap_all['COMISION_TOTAL']
swap_all.head()
```

### SWAP BY ESTATUS

```python
swap_status = all_respuestas.groupby(['Metodo de dispersion','Estatus','file_name']).agg({'Importe':sum,'report_id':'count'}).reset_index()
swap_status['COMISION'] = 1.85
swap_status['IVA'] = swap_status['COMISION'] * 0.16
swap_status['COMISION + IVA'] = swap_status['COMISION'] + swap_status['IVA']
swap_status['COMISION_FINAL'] = swap_status['COMISION'] * swap_status['report_id']
swap_status['IVA_FINAL'] = swap_status['COMISION_FINAL'] * 0.16
swap_status['COMISION_CALCULADA'] = swap_status['COMISION_FINAL'] + swap_status['IVA_FINAL']
swap_status['COMISION_TOTAL'] = swap_status['report_id'] * swap_status['COMISION + IVA']
swap_status['TOTAL_DEPOSITADO'] = swap_status['Importe'] + swap_status['COMISION_TOTAL']
swap_status.head()
```

```python
swap_status['Estatus'].unique()
```

# EXCEL

```python
writer = pd.ExcelWriter('Historico 2023.xlsx',
                        engine='xlsxwriter',
                        engine_kwargs={'options': {'encoding':'utf-8'}})
swap_all.to_excel(writer, sheet_name='SWAP_ALL',index=False)
swap_status.to_excel(writer, sheet_name='SWAP_STATUS',index=False)
writer.close()
```
