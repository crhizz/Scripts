```python
import pandas as pd
import numpy as np
import glob
import os
from openpyxl import load_workbook
pd.options.display.float_format = '{:,.2f}'.format
```

```python
start_date = '2024-01-01'
final_date = '2024-01-31'
```

```python
trad_files = glob.glob(r'C:\Users\Cristian.Aparicio\Documents\DISPERSIONES\2024\01. ENERO\ARCHIVO DISPERSIONES\Dispersión*.xlsx', recursive=False)
trad_files
```

# SWAP DEL DÍA

```python
sf_final = pd.concat([pd.read_excel(f,sheet_name='Swap del día',
                                          converters={'CLABE':str,'report_id':str}).assign(file_name=os.path.basename(f)) for f in trad_files])
sf_final['date'] = sf_final["file_name"].str[-13:-5]
sf_final['date']= pd.to_datetime(sf_final['date'], format='%Y%m%d').dt.date
sf_final['date'] = pd.to_datetime(sf_final['date'])
sf_final['date_index'] = sf_final['date'].astype(str)
sf_final['report_date_local']= pd.to_datetime(sf_final['report_date_local'])
sf_final['INDEX'] = sf_final['report_id'] + '_' + sf_final['date_index']
sf_final.head()
```

```python
sf_final.shape
```

# DEVOLUCIONES

```python
dev_total = pd.concat([pd.read_excel(f,sheet_name='ERROR').assign(file_name=os.path.basename(f)) for f in trad_files])
dev_total['date'] = dev_total["file_name"].str[-13:-5]
dev_total['date']= pd.to_datetime(dev_total['date'], format='%Y%m%d').dt.date
dev_total['date']= pd.to_datetime(dev_total['date'])
dev_total['report_date_local']= pd.to_datetime(dev_total['report_date_local'])
dev_total.head()
```

```python
dev_total.shape
```

```python
dev_total['tipo'].unique()
```

### DEVOLUCIONES TOTALES

```python
dev_total_pivot = pd.pivot_table(dev_total,index=["date"],values=['total_due'],aggfunc = ['sum', len],margins=True)
dev_total_pivot.reset_index(drop=False, inplace=True)
dev_total_pivot.head()
```

```python
dev_total_pivot.shape
```

### DEVOLUCIONES PAGO DEL DIA 

```python
dev_dia = dev_total[(dev_total['tipo'] == 'Pago del día')]
```

```python
dev_dia_pivot = pd.pivot_table(dev_dia,index=["date"],values=['total_due'],aggfunc = ['sum', len],margins=True)
dev_dia_pivot.reset_index(drop=False, inplace=True)
dev_dia_pivot.head()
```

```python
dev_dia_pivot.shape
```

### DEVOLUCIONES REPROGRAMACION

```python
dev_repro = dev_total[(dev_total['tipo'] == 'Reprogramación')]
```

```python
dev_repro_pivot = pd.pivot_table(dev_repro,index=["date"],values=['total_due'],aggfunc = ['sum', len],margins=True)
dev_repro_pivot.reset_index(drop=False, inplace=True)
dev_repro_pivot.head()
```

```python
dev_repro_pivot.shape
```

## EXCEL DEVOLUCIONES

```python
writer = pd.ExcelWriter('ERROR ENERO 2024.xlsx',
                        engine='xlsxwriter',
                        engine_kwargs={'options': {'encoding':'utf-8'}})
dev_total_pivot.to_excel(writer, sheet_name='TOTAL ERROR')
dev_dia_pivot.to_excel(writer, sheet_name='PAGO DEL DIA')
dev_repro_pivot.to_excel(writer, sheet_name='REPROGRAMACION')
writer.close()
```

### PIVOT BNTE

```python
respuestas_final_bnte = respuestas_final[(respuestas_final['Metodo de dispersion'] == 'BANORTE')]
respuestas_final_bnte.head()
```

```python
respuestas_final_bnte.shape
```

```python
def error_final_bnte(respuestas_final_bnte):
    if respuestas_final_bnte['Estatus'] == 'PAID':
        return 'PAID'
    else:
        return 'ERROR'
    
respuestas_final_bnte['Estatus_final'] = respuestas_final_bnte.apply(error_final_bnte, axis=1)
```

```python
pt_bnte = pd.pivot_table(respuestas_final_bnte,index=['Date_respuesta','Estatus_final']
                         ,values=["Importe",'COMISION',"COMISION SIN IVA",'IVA']
                         ,aggfunc=np.sum)
pt_bnte = pt_bnte.reset_index(drop=False)
pt_bnte.head()
```

```python
pt_bnte.shape
```

### PIVOT SANTANDER

```python
respuestas_final_sant = respuestas_final[(respuestas_final['Metodo de dispersion'] == 'SANTANDER')]
respuestas_final_sant.head()
```

```python
respuestas_final_sant.shape
```

```python
def error_final_sant(respuestas_final_sant):
    if respuestas_final_sant['Estatus'] == 'PAID':
        return 'PAID'
    else:
        return 'ERROR'
    
respuestas_final_sant['Estatus_final'] = respuestas_final_sant.apply(error_final_sant, axis=1)
```

```python
pt_sant = pd.pivot_table(respuestas_final_sant,index=['Date_respuesta','Estatus_final']
                         ,values=["Importe",'COMISION',"COMISION SIN IVA",'IVA']
                         ,aggfunc=np.sum)
pt_sant = pt_sant.reset_index(drop=False)
pt_sant.head()
```

```python
pt_sant.shape
```
