# PAGOS NEW DE MESES ANTERIORES

```python
query_new_last_month = query_error[(query_error['mes_cierre'] != 'Actual') & (query_error['status_cubeta'] == 'NEW')]
query_new_last_month.sort_values(by=['day'],inplace=True)
query_new_last_month.head()
```

```python
query_new_last_month.shape
```

# IDENTIFICACIÓN DE PAGOS

```python
cruce_1 = pd.concat([query_error_current_month,query_new_last_month])
cruce_1['month'].fillna('Not found',inplace=True)
cruce_1.head()
```

```python
cruce_1.shape
```

```python
def tipo_pago(cruce_1):
    if cruce_1['report_date_local'] == cruce_1['ultimo_mov']:
        return 'Pago del día'
    else:
        return 'Reprogramación'
cruce_1['tipo_pago'] = cruce_1.apply(tipo_pago, axis=1)
```

```python
duplicates_cruce_1 = cruce_1[cruce_1.duplicated(['report_id'] , keep=False)]
duplicates_cruce_1.head()
```

# AGRUPADO POR TIPO DE PAGO - MES ACTUAL

```python
group_tipo_pago = pd.pivot_table(cruce_1,index=['ultimo_mov_date_only'],
                                 columns=['esquema_query','status_cubeta','tipo_pago'],
                                 values=['amount'],
                                 aggfunc=np.sum,margins=True).reset_index()
group_tipo_pago.fillna(0,inplace=True)
group_tipo_pago.head()
```
